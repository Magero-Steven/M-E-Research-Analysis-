## Importance of loops
Loops are an important concept in programming ; they are essential in automating repetitive tasks and handling large datasets.


Imagine having alist of trees and required to print the name of every tree

```python
tree_list=["Acacia", "Baobab","Cedar","Cypress"]
print("Acacia")
print("Baobab")
print("Cedar")
print("Cypress")
print("Cypress")
```

```python
tree_list.extend(["Eucalyptus","Fig","Jacaranda","Mango","Mahogany","Neem",
                  "Oak","Olive","Palm","Pine","Sycamore","Teak","Willow"])
len(tree_list)
```

It is very manageable for  a smaller list but very  tideos for a larger list. Thats where loops come in.

```python
for i in tree_list:
  print(i)
```

## `for` loops

 **`for` loops:** Ideal for going through a sequence (like a list, tuple, or string), one by one. We use `for` loops when we know how many things there are in a list.

    
    For example, for each element `tree` in a list of trees `tree_list2 `, print out the value of `tree`:

```python
tree_list2=["Acacia", "Baobab","Cedar","Cypress"]
for tree in tree_list2:
  print (tree)
```

## `while` loops

* **`while` loops:** While loops continue to execute a set of instructions as long as a certain condition is met.
    
    For example, while `i` is smaller than `5`, print out the value of `i` and add 1 to it.

```python
i = 0

while i < 5:
    print(i)
    i += 1
```

**`caution`** ;Be cautious with `while` loops, because they result in infinite loops if the condition never becomes false. Always ensure there's a way for the condition to become false eventually.
